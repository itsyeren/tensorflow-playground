<details>
    <summary>Answer: some insights about activation functions</summary>

- Since ReLU is piecewise linear, the results may look hexagonal!
- A linearly non-separable problem cannot be fitted with a linear activation like **Linear**
- Surprisingly, a piecewise linear activation function like **ReLU** (or **LeakyReLU**) fits this linearly non-separable problem well (though not always)
- The `tanh` activation provides a "smoother" decision boundary
- **Sigmoid** doesn't seem to work well here (i.e., it requires __way__ too many epochs to classify effectively)

🧑🏻‍🏫 Always start with ReLU, it's a safe choice 🧑🏻‍🏫!

### (4) Building and Fitting a Neural Network in ***Tensorflow.Keras***

👇 We wrote the same model for you in Tensorflow's Keras — at least the architecture.

````python
from keras import Sequential, Input, layers

model = Sequential()

model.add(Input(shape=(2,))) # input layer with 2 features

model.add(layers.Dense(5, activation='relu')) # 1st hidden layer with 5 neurons
model.add(layers.Dense(4, activation='relu')) # 2nd hidden layer with 4 neurons
model.add(layers.Dense(3, activation='relu')) # 3rd hidden layer with 3 neurons

model.add(layers.Dense(1, activation='sigmoid')) # output layer giving the probability of belonging
                                                 # to the "positive" class
````

<details>
  <summary><i>What should we understand about this Neural Network code?</i> 👆</summary>

- <u>First Hidden Layer (a.k.a. ***Input Layer***)</u>:
  - Each data point entering the neural network has two features:  
    \( X = \begin{bmatrix} X_{1} \\ X_{2} \end{bmatrix} \)
  - You need to specify the ***number of input features*** to the neural network via the ***`input_dim` parameter***.
  - A neural network tries to mimic the human brain. Here, we want to use 5 neurons to start analyzing these points.
  - Each data point passes through the first hidden layer made up of 5 neurons:  
    \( layer_1 = \begin{bmatrix} a_{1} \\ a_{2} \\ a_{3} \\ a_{4} \\ a_{5} \end{bmatrix} \)

- <u>Second Hidden Layer</u>:
  - What if we want information to ***flow*** through a second hidden layer with 4 neurons? That is entirely possible!
  - These 4 neurons in the second layer  
    \( layer_2 = \begin{bmatrix} b_{1} \\ b_{2} \\ b_{3} \\ b_{4} \end{bmatrix} \),  
    analyze the output of the 5 neurons in the first layer.

- <u>Third Hidden Layer</u>:
  - What if we want information to **keep flowing** through a third hidden layer with 3 neurons? Also entirely possible!
  - The output of each neuron in the second layer passes through the third hidden layer made up of 3 neurons:  
    \( layer_3 = \begin{bmatrix} c_{1} \\ c_{2} \\ c_{3} \end{bmatrix} \)
  - These 3 neurons analyze the output of the neurons in \( layer_2 \).

- <u>Predictive Layer</u>:
  - You are dealing with a binary classification problem.
  - Two neurons could be used to predict the probability of belonging to class A or B...
  - But a single neuron predicting the probability of "success" is sufficient.

- <u>About activation functions</u>:
  - Despite its simplicity, ***ReLU*** is very effective at adding non-linearity to the layers.
  - For the predictive layer, the best activation function to use for classification problems is the ***sigmoid*** function.  
    This was discussed previously in the Decision Science and Machine Learning sections.

- <u>About the Sequential structure of the network</u>:
  - A consequence of defining a **Sequential** model is that  
    each layer automatically knows its own input size by looking at the output size of the previous layer.

</details>
